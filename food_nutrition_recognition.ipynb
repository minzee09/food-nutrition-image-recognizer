```python
# 데이터 전처리 함수 정의
def preprocess_image_with_nutrition(image, label):
    print("Image shape:", image.shape)
    print("Label:", label)
    # Resize
    image = tf.image.resize(image, [224, 224])
    image = tf.cast(image, tf.float32) / 255.0

   #라벨 텐서 -> 파이썬 문자열 변환
    label_str = tf.strings.as_string(label)
    label_str_np = label_str.numpy().decode('utf-8')
    nutritional_info = nutrition_mapping.get(label_str_np, {'sugar': 0, 'carbs': 0, 'protein': 0, 'fat': 0})

    sugar = tf.constant(nutritional_info['sugar'], dtype=tf.float32)
    carbs = tf.constant(nutritional_info['carbs'], dtype=tf.float32)
    protein = tf.constant(nutritional_info['protein'], dtype=tf.float32)
    fat = tf.constant(nutritional_info['fat'], dtype=tf.float32)

    nutrition_tensor = tf.stack([sugar, carbs, protein, fat])

    return image, nutrition_tensor
```
